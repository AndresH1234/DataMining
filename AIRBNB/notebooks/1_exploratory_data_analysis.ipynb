## Valores Nulos

```python
# Cambiar los nulos por -1
df["bathrooms"] = df["bathrooms"].fillna(-1)
```

```python
# Revisar los nulos de host_has_profile_pic
df[df["host_has_profile_pic"].isnull()]
```

```python
# Ver los valores Ãºnicos de la columna host_has_profile_pic
df["host_has_profile_pic"].unique()
```

```python
# Ver si en los nulos de host_has_profile_pic hay algÃºn valor que no sea NaN en la columna host_picture_url
df[df["host_has_profile_pic"].isnull() & df["host_identity_verified"].notnull() & df["host_since"].notnull()]
```

```python
# Ver los valores unicos de host_identity_verified
print("host_has_profile_pic",df["host_has_profile_pic"].unique())
print("host_identity_verified",df["host_identity_verified"].unique())
print("host_since",df["host_since"].unique())
```

```python
# Cambiar los nulos por -1
df["host_has_profile_pic"] = df["host_has_profile_pic"].fillna(-1)
```

```python
df["host_identity_verified"] = df["host_identity_verified"].fillna(-1)
```

```python
# Cambiar los nulos de host_since por 0000-00-00
df["host_since"] = df["host_since"].fillna("0000-00-00")
```

```python
# ver los nulos de host response rate
df[df["host_response_rate"].isnull()]
```

```python
# Ver los valores Ãºnicos de host_response_rate
df["host_response_rate"].unique()
```

```python
# Cambiar los nulos por -1
df["host_response_rate"] = df["host_response_rate"].fillna(-1)
```

```python
# pasar de 20% a 0.20
df["host_response_rate"] = df["host_response_rate"].str.replace("%","").astype(float) / 100
```

```python
# ver los nulos de review_scores_rating
df[df["review_scores_rating"].isnull()].head()
```

```python
# Ver si hay algÃºn valor en review_scores_rating que no sea NaN con number_of_reviews 0
df[(df["review_scores_rating"].isnull()) & (df["number_of_reviews"] != 0) & (df["number_of_reviews"] != 1)]
```

```python
# Ver los valores Ãºnicos de review_scores_rating
df["review_scores_rating"].unique()
```

```python
# Reemplazar los nulos por -1
df["review_scores_rating"] = df["review_scores_rating"].fillna(-1)
```

```python
# Ver los nulos de bedrooms
df[df["bedrooms"].isnull()]
```

```python
# Ver los valores Ãºnicos de bedrooms
df["bedrooms"].unique()
```

```python
# Confirmar si existen valores decimales
df["bedrooms"].dropna().apply(lambda x: x.is_integer()).unique()
```

```python
# Reemplazar los nulos por -1
df["bedrooms"] = df["bedrooms"].fillna(-1)
```

```python
# Ver nulos en beds
df[df["beds"].isnull()]
```

```python
# ver los valores Ãºnicos de beds
df["beds"].unique()
```

```python
df['beds'] = df['beds'].fillna(-1)
```

```python
df["beds"].dropna().apply(lambda x: x.is_integer()).unique()
```

```python
# Transforma la columan beds y bathrooms a entero
df["bedrooms"] = df["bedrooms"].astype(int)
df["beds"] = df["beds"].astype(int)
```

```python
df["beds"].unique()
```
